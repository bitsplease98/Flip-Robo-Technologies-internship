```python
import selenium
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from tqdm.notebook import tqdm_notebook
import pandas as pd
```

```python
# Empty list to store data
purchase_year = []     # store all data in a list
owner = []         # store number of owner
history = []       # store history of accident
transmission = []  # store transmission type
fuel_type = []     # store fuel type
km_driven = []     # store km driven
last_service = []  # store last servicing
registration = []  # store registration
model = []         # store model of car
price = []         # store price of a car
```

```python
def car24_data(url):
    
    # calling the webdriver and then providing the desired url
    driver = webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    driver.maximize_window()

    # scroll down the page to the table
    SCROLL_PAUSE_TIME = 0.5

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    # empty list to store urls
    car24_urls=[]
    for i in driver.find_elements_by_xpath("//a[@class='_9Ue0B']"):
        car24_urls.append(i.get_attribute('href'))
    
    print(len(car24_urls))
    

    # setting a loop from the urls list to find data
    for i in tqdm_notebook(car24_urls[2:]):
        driver.get(i)
        driver.implicitly_wait(10)
        
        # scrapping model of car
        try:
            model.append(driver.find_element_by_xpath('//h2[@class="_3FpCg"]').text)
        except NoSuchElementException:
            model.append("Not Available")
        
        # scraping accident history of the car
        try:
            history.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div/div/ul/li[1]/strong').text)
        except NoSuchElementException:
            history.append("Not Available")
            
        # Scraping year of purchase
        try:
            purchase_year.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[9]/strong').text)
        except NoSuchElementException:
            purchase_year.append("Not Available")
            
        # Scraping number of owner
        try:
            owner.append(driver.find_element_by_css_selector('body > div.js-content > div._3Ii-d > div._1AvBW > div > div > div > div > ul > li:nth-child(2) > strong').text)
        except NoSuchElementException:
            owner.append("Not Available")
            
        # Scraping transmission type
        try:
            transmission.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[6]/strong').text)
        except NoSuchElementException:
            transmission.append("Not Available")
            
        # Scraping fuel type
        try:
            fuel_type.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[4]/strong').text)
        except NoSuchElementException:
            fuel_type.append("Not Available")
            
        # Scraping km driven
        try:
            km_driven.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[3]/strong').text)
        except NoSuchElementException:
            km_driven.append("Not Available")
            
        # Scraping last service
        try:
            last_service.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[5]/strong').text)
        except NoSuchElementException:
            last_service.append("Not Available")
        
        # Scraping registration
        try:
            registration.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div[3]/div/div/div[1]/div/ul/li[7]/strong').text)
        except NoSuchElementException:
            registration.append("Not Available")
        
        
        # scrapping price of car
        try:
            price.append(driver.find_element_by_xpath('//div[@class="_7udZZ"]/span').text)
        except NoSuchElementException:
            price.append("Not Available") 
```

#### Scraping car details for the city New Delhi

```python
# Calling function

car24_data("https://www.cars24.com/buy-used-car?sort=P&storeCityId=2&pinId=110001")
```

```python
print(len(model),len(transmission),len(fuel_type),len(purchase_year),len(last_service),
      len(registration),len(km_driven),len(owner),len(price))
```

```python
# Saving data into dataframe

car24_data=pd.DataFrame({"Model":model,"History":history,"Transmission Type":transmission,"Fuel Type":fuel_type,
                         "Purchase Year":purchase_year,"Last Service":last_service,"Registration":registration,
                         "KM Driven":km_driven,"Owner":owner,"Price":price})

car24_data
```

```python
# Saving the excel file

car24_data.to_excel(r"E:\Flip Robo\Project - 4 - Car Price Pridiction\Car Details\car24_delhi.xlsx")
```
