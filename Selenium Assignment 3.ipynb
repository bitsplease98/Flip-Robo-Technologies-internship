### Q1. Scrape the details of most viewed videos on YouTube from Wikipedia:
##### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
##### You need to find following details:
##### A) Rank
##### B) Name
##### C) Artist
##### D) Upload date
##### E) Views

```python
def most_viewed(u):

    driver=webdriver.Edge(r'C:\Users\aush1\Downloads\msedgedriver.exe')
    driver.get(u)
    time.sleep(2)

    rank=[]
    name=[]
    artist=[]
    dat=[]
    views=[]

    #Rank
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[1]'):
        rank.append(i.text)
    #Name
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[2]'):
        name.append(i.text)
    #artist
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[3]'):
        artist.append(i.text)
    #Date
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[5]'):
        dat.append(i.text)
    #Views
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter" ][1]/tbody/tr/td[4]'):
        views.append(i.text)

    #Creating Dataframe
    data=pd.DataFrame()
    data['Rank']=rank
    data['Name']=name
    data['Artist']=artist
    data['Date']=dat
    data['Views']=views
    return data
```

### Q2. Scrape the details team India’s international fixtures from bcci.tv.
##### Url = https://www.bcci.tv/.
##### You need to find following details:
##### A) Match title (I.e. 1st ODI)
##### B) Series
##### C) Place
##### D) Date
##### E) Time
##### Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
def fixtures(u):

    driver=webdriver.Edge(r'C:\Users\aush1\Downloads\msedgedriver.exe')
    driver.get(u)
    driver.maximize_window()

    inter=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]')
    fix=driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a')
    action=ActionChains(driver)
    action.move_to_element(inter).move_to_element(fix).click().perform()
    time.sleep(2)

    title=[]
    series=[]
    place=[]
    date=[]
    t=[]

    #Title
    for i in driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'):
        title.append(i.text)

    #Series
    for i in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__tournament-label u-truncated"]'):
        series.append(i.text)

    #Place
    for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]/span'):
        place.append(i.text)

    #Date

    day=[]
    month=[]

    #Extracting day
    for i in driver.find_elements_by_xpath('//span[@class="fixture__date"]'):
        day.append(i.text)

    #Extracting month
    for i in driver.find_elements_by_xpath('//span[@class="fixture__month"]'):
        month.append(i.text)

    #joining day and month into one column
    for i in range(len(day)):
        date.append(day[i]+month[i])

    #Time
    for i in driver.find_elements_by_xpath('//span[@class="fixture__time"]'):
        t.append(i.text)


    #Creating Dataframe
    data=pd.DataFrame()
    data['Match Title']=title
    data['Series']=series
    data['Place']=place
    data['Date']=date
    data['Time']=t
    return data
```

### Q5. Scrape the details of trending repositories on Github.com.
##### Url = https://github.com/
##### You have to find the following details:
##### A) Repository title
##### B) Repository description
##### C) Contributors count
##### D) Language used
##### Note: - From the home page you have to click on the trending option from Explore menu through code.


```python
def git(u):

    driver=webdriver.Edge(r'C:\Users\aush1\Downloads\msedgedriver.exe')
    driver.get(u)
    driver.maximize_window()

    f=0
    #Looping till element becomes interactable
    while f==0:
        exp=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary')
        trend=driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a')
        action=ActionChains(driver)
        action.move_to_element(exp).click().perform()
        time.sleep(2)
        try:
            action.move_to_element(trend).click().perform()
            f=1
        except ElementNotInteractableException:
            pass

    title=[]
    url=[]
    desc=[]
    contri=[]
    lang=[]

    #Repository Title and url for each of these repositories
    for i in driver.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
        title.append(i.text)
        url.append(i.get_attribute('href'))
    #For each URL
    for j in url:
        driver.get(j)
        #Repository Description
        try:
            d = driver.find_element_by_xpath('//p[@class="f4 mt-3"]')
            desc.append(d.text)
        except NoSuchElementException:
            desc.append('-')
        #Contribution Count
        try:
            contri.append(driver.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[4]/div/h2/a/span').text)
        except NoSuchElementException:
            try:
                contri.append(driver.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[5]/div/h2/a/span').text)
            except NoSuchElementException:
                contri.append('-')
        #Language Used
        s=''
        try:
            for i in driver.find_elements_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[5]/div/ul/li'):
                s=s+'/'+i.text.replace('\n',' ').split()[0]
        except:
            for i in driver.find_elements_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[6]/div/ul/li'):
                s=s+'/'+i.text.replace('\n',' ').split()[0]
        lang.append(s)

    #Creating Dataframe
    data=pd.DataFrame()
    data['Repository Title']=title
    data['Repository Description']=desc
    data['Contribution Count']=contri
    data['Language Used']=lang
    return data
```

```python
df=git('https://github.com/')
df
```

### Q6. Scrape the details of top 100 songs on billboard.com.

### Q7. Scrape the details of Data science recruiters from naukri.com.
##### Url = https://www.naukri.com/
##### You have to find the following details:
##### A) Name
##### B) Designation
##### C) Company
##### D) Skills they hire for
##### E) Location
##### Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
def naukri(u):
    
    driver=webdriver.Edge(r'C:\Users\aush1\Downloads\msedgedriver.exe')
    driver.get(u)
    driver.maximize_window()

    #Getting recruiter link and providing it to the driver
    rec=driver.find_element_by_xpath('//*[@id="root"]/div[1]/div/ul[1]/li[2]/a')
    driver.get(rec.get_attribute('href'))

    #Searching for Data Scientist 
    search=driver.find_element_by_xpath('//*[@id="skill"]/div[1]/div[2]/input')
    search.send_keys('Data Science')
    time.sleep(1)
    search.send_keys(Keys.ENTER)

    name=[]
    desig=[]
    comp=[]
    skills=[]
    loc=[]
    url=[]

    #Name
    for i in driver.find_elements_by_xpath('//span[@class="fl ellipsis"]'):
        name.append(i.text)
    #Designation
    for i in driver.find_elements_by_xpath('//span[@class="ellipsis clr"]'):
        desig.append(i.text)
    #Company
    for i in driver.find_elements_by_xpath('//a[@class="ellipsis"][2]'):
        comp.append(i.text)
    #Skills
    for i in driver.find_elements_by_xpath('//div[@class="hireSec highlightable"]'):
        skills.append(i.text)
    #URL
    for i in driver.find_elements_by_xpath('//a[@class="ellipsis"][1]'):
        url.append(i.get_attribute('href'))

    #For every url we are extracting location
    for u in url:
        driver.get(u)
        WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div[1]/div[1]/div[2]/div[5]/a')))
        loc.append(driver.find_element_by_xpath('/html/body/div[2]/div[1]/div[1]/div[2]/div[5]/a').text)

    #Creating Dataframe
    data=pd.DataFrame()
    data['Name']=name
    data['Designation']=desig
    data['Company']=comp
    data['Skills they higher for']=skills
    data['Location']=loc
    return data
```

```python
df=naukri('https://www.naukri.com/')
df
```

### Q9. Scrape the details most watched tv series of all time from imdb.com.

##### You have to find the following details:
##### A) Name
##### B) Year span
##### C) Genre
##### D) Run time
##### E) Ratings
##### F) Votes

```python
def tv_series(u):

    driver=webdriver.Edge(r'C:\Users\aush1\Downloads\msedgedriver.exe')
    driver.get(u)
    driver.maximize_window()

    name=[]
    year=[]
    gen=[]
    rt=[]
    rat=[]
    vot=[]

    #Name
    for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a'):
        name.append(i.text)
    #Year Span
    for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/span[2]'):
        year.append(i.text)
    #Genre
    for i in driver.find_elements_by_xpath('//span[@class="genre"]'):
        gen.append(i.text)
    #Run TIme
    for i in driver.find_elements_by_xpath('//span[@class="runtime"]'):
        rt.append(i.text)
    #Ratings
    for i in driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]/span[2]'):
        rat.append(i.text)
    #Votes
    for i in driver.find_elements_by_xpath('//span[@name="nv"]'):
        vot.append(i.text)

    #Creating Dataframe
    data=pd.DataFrame()
    data['Name']=name
    data['Year Span']=year
    data['Genre']=gen
    data['Run Time']=rt
    data['Ratings']=rat
    data['Votes']=vot
    return data
```
