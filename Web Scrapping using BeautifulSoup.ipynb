### Q1.Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’

```python
def headertags(u):
    head=[]
    page=requests.get(u)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        headers=soup.find_all(['h{}'.format(i) for i in range(1,7)])
        for i in headers:
            s=i.text.replace('\n','')
            head.append(s.replace('\xa0',''))
        return head
    else:
        return('No permission for scrapping the web page')
```

### Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file. 

```python
def top100(u):
    name=[]
    rating=[]
    year=[]
    page=requests.get('https://www.imdb.com/chart/top')
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        #Scrapping names of the movies
        n=soup.find_all('td',class_='titleColumn')
        k=0
        for i in n:
            for j in i.find_all('a'):
                name.append(j.text)
            k+=1
            if k==100:
                break
        #Scrapping ratings     
        r=soup.find_all('td',class_='ratingColumn imdbRating')
        k=0
        for i in r:
            rating.append(float(i.text.replace('\n','')))
            k+=1
            if k==100:
                break
        #Scrapping Year       
        y=soup.find_all('span',class_='secondaryInfo')
        k=0
        for i in y:
            s=i.text.replace('(','')
            year.append(int(s.replace(')','')))
            k+=1
            if k==100:
                break
    else:
        return('No permission for scrapping the web page')
    #Converting all data into Dataframe and returning it
    imdb=pd.DataFrame()
    imdb['Movie']=name
    imdb['Ratings']=rating
    imdb['Year']=year
    return imdb
```

```python
df=top100('https://www.imdb.com/chart/top')
df
```

```python
df.to_csv('top100imdb.csv',index=False)
```

### Q4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def books(u):
    name=[]
    auth=[]
    gen=[]
    rev=[]
    page=requests.get(u)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        
        #Scrapping book names
        count=0
        n=soup.find_all('h4',class_='italic')
        for i in n:
            s=i.text.replace('\n','')
            name.append(s.replace('★','').strip())
            count+=1
            if count==5:
                break
                
        #Scrapping Book authors
        count=0
        a=soup.find_all('p',class_='sans bold')
        for i in a:
            auth.append(i.text.replace('\n',''))
            count+=1
            if count==5:
                break
                
        #Scraping Genre
        count=0
        g=soup.find_all('p',class_='genre-links hidden-phone')
        for i in g:
            s=''
            for j in i('a'):
                s+=j.text+'/'
            gen.append(s[:-1])
            count+=1
            if count==5:
                break
        #Scraping Review
        count=0
        r=soup.find_all('p')
        for i in r:
            if count%4==0 and count>0:
                rev.append(i.text.replace('\n',''))
            count+=1
            if count==21:
                break
    else:
        print('No Permission for scrapping')
    
    #Creating dataframe
    df=pd.DataFrame()
    df['Book name']=name
    df['Author']=auth
    df['Genre']=gen
    df['Review']=rev
    return df
```

```python
data=books('https://bookpage.com/reviews')
data
```

### Q5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:

##### i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
def oditeams(u):
    team=[]
    match=[]
    point=[]
    rating=[]
    page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        #Scrapping Team Names
        count=0
        t=soup.find_all('span',class_='u-hide-phablet')
        for i in t:
            team.append(i.text)
            count+=1
            if count==10:
                break
        #Scrapping Matches and Points
        count=0
        m1=soup.find('td',class_='rankings-block__banner--matches')
        match.append(m1.text)
        p1=soup.find('td',class_='rankings-block__banner--points')
        point.append(p1.text)
        mp=soup.find_all('td',class_='table-body__cell u-center-text')
        for i in mp:
            if count%2==0:
                match.append(i.text)
            else:
                point.append(i.text)
            count+=1
            if count==18:
                break
        #Scrapping Ratings        
        count=0
        r1=soup.find('td',class_='rankings-block__banner--rating u-text-right')
        rating.append(r1.text.replace('\n','').strip())
        r=soup.find_all('td',class_='table-body__cell u-text-right rating')
        for i in r:
            rating.append(i.text)
            count+=1
            if count==9:
                break
        else:
            print('No Permission for Scrapping')
        #COnvering it into a Dataframe
        df=pd.DataFrame()
        df['Teams']=team
        df['Matches']=match
        df['Points']=point
        df['Ratings']=rating
        return df    
```

##### ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 

```python
def batsmen(u):
    batsmen=[]
    teams=[]
    ratings=[]
    page=requests.get(u)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        
        #Scrapping info of the first player
        b1=soup.find('div',class_='rankings-block__banner--name')
        t1=soup.find('div',class_='rankings-block__banner--nationality')
        r1=soup.find('div',class_='rankings-block__banner--rating')
        batsmen.append(b1.text)
        teams.append(t1.text.strip()[:3])
        ratings.append(r1.text)
        
        #Scrapping all the batsmen names
        count=0
        b=soup.find_all('td',class_='table-body__cell name')
        for i in b:
            batsmen.append(i.text.replace('\n',''))
            count+=1
            if count==9:
                break
                
        #Scrapping all the teams name
        count=0
        t=soup.find_all('span',class_='table-body__logo-text')
        for i in t:
            teams.append(i.text)
            count+=1
            if count==9:
                break
                
        #Scrapping all the ratings
        count=0
        r=soup.find_all('td',class_='table-body__cell u-text-right rating')
        for i in r:
            ratings.append(i.text)
            count+=1
            if count==9:
                break
    else:
        print('No permission to scrape')
        
    #Storing in Dataframe
    df=pd.DataFrame()
    df['Batsmen']=batsmen
    df['Team']=teams
    df['Rating']=ratings
    return df
    
```

### Q6.Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:


##### i) Top 10 ODI teams in women’s cricket along with the records for matches, points and rating. 


```python
def womenodi(u):
    name=[]
    match=[]
    point=[]
    rating=[]
    page=requests.get(u)
    if page.status_code==200:
        soup=BeautifulSoup(page.content)
        
        #Scrapping info of first team
        n1=soup.find('span',class_='u-hide-phablet')
        n1=n1.text.replace('[','')
        m1=soup.find('td',class_='rankings-block__banner--matches')
        m1=m1.text.replace('[','')
        p1=soup.find('td',class_='rankings-block__banner--points')
        p1=p1.text.replace('[','')
        r1=soup.find('td',class_='rankings-block__banner--rating u-text-right')
        name.append(n1.replace(']',''))
        match.append(m1.replace(']',''))
        point.append(p1.replace(']',''))
        rating.append(r1.text.replace('\n','').strip())
        
        #Scraping info of rest of the teams
        x=soup.find_all('tr',class_='table-body')
        count=0
        for i in x:
            for n in i.find_all('span',class_='u-hide-phablet'):
                name.append(n.text)
            for m in i.find_all('td',class_='table-body__cell u-center-text'):
                if count%2==0:
                    match.append(m.text)
                else:
                    point.append(m.text)
                count+=1
            for r in i.find_all('td',class_='table-body__cell u-text-right rating'):
                rating.append(r.text)
    else:
        print('No permission to scrap')
    
    #Creating dataframe
    df=pd.DataFrame()
    df['Team name']=name
    df['Matches played']=match
    df['Points']=point
    df['Ratings']=rating
    return df
        
```

### Q7.Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def phonesunder20k(link):
    name=[]
    price=[]
    img=[]
    rating=[]
    for i in range(1,21):
        u=link.format(i)
        page=requests.get(u)
        soup=BeautifulSoup(page.content)
        
        # Phone name
        m=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
        for i in m:
            name.append(i.text)
        
        #price
        p=soup.find_all('div',class_='sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20')
        for i in p:
            for j in i.find_all('span',class_='a-price-whole'):
                if j.text!='':
                    price.append(j.text)
                else:
                    price.append(np.NaN)
                
        #image url of the phone
        im=soup.find_all('div',class_='a-section aok-relative s-image-fixed-height')
        for i in im:
            for j in i.find_all('img'):
                if 'src' in j.attrs:
                    if j.attrs!='':
                        img.append(j.attrs['src'])
                    else:
                        img.append(np.NaN)

                        
        #ratings
        r=soup.find_all('span',class_='a-icon-alt')
        for i in r:
            rate=i.text
            if rate!='':
                rating.append(rate)
            else: 
                rating.append(np.NaN)
    
    m=min(len(name),len(price),len(img),len(rating))
    
    #creating a dataframe
    df=pd.DataFrame()
    df['Mobile']=name[:m]
    df['Price']=price[:m]
    df['Image Url']=img[:m]
    df['Ratings']=rating[:m]
    return df
```

```python
df=phonesunder20k('https://www.amazon.in/s?k=phone+under+20000&page={}&qid=1621869971&ref=sr_pg_2')
df
```

### Q8.Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description.

```python
def weather(u):
    period=[]
    sdesc=[]
    temp=[0]*9
    templow=[]
    temphigh=[]
    desc=[]
    page=requests.get(u)
    soup=BeautifulSoup(page.content)
    x=soup.find_all('div',id='seven-day-forecast-container')

    for days in x:
        for j in days.find_all('div',class_='tombstone-container'):
            for p in j.find_all('p',class_='period-name'):   #period
                period.append(p.text)
            for s in j.find_all('p',class_='short-desc'):    #short description
                sdesc.append(s.text)
            for tlow in j.find_all('p',class_='temp temp-low'):    #low temperature
                templow.append(tlow.text)
            for thigh in j.find_all('p',class_='temp temp-high'):    #high temperature
                temphigh.append(thigh.text)
    
    #description            
    x=soup.find_all('div',class_='col-sm-10 forecast-text')
    count=0
    for i in x:
        desc.append(i.text)
        count+=1
        if count==9:
            break
            
    #converting high temp lis and low temp list into one list --> temp
    for i in range(len(templow)):
        temp[i*2]=templow[i]
    c=1
    for i in range(len(temphigh)):
        temp[i+c]=temphigh[i]
        c+=1
        
    #creating a dataframe
    df=pd.DataFrame()
    df['Period']=period
    df['Tempertature']=temp
    df['Short Description']=sdesc 
    df['Description']=desc 
    return df
    
    
```

```python
df=weather('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YKvYZY3itPY')
df
```

### Q9.Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date

```python
def fresherjobs(u):
    title=[]
    company=[]
    ctc=[]
    date=[]
    page=requests.get(u.format('/'))
    soup=BeautifulSoup(page.content)
    
    #Getting total number of pages
    p=soup.find('span',id='total_pages')
    total_pages=int(p.text)
    
    for p in range(1,total_pages+1):
        page=requests.get(u.format('/page-'+str(p)))
        soup=BeautifulSoup(page.content)
        
        
        #Job title
        findtitle=soup.find_all('div',class_='heading_4_5 profile')
        for i in findtitle:
            title.append(i.text.replace('\n',''))
         
        
        #Company
        comp=soup.find_all('div',class_='heading_6 company_name')
        for i in comp:
            company.append(i.text.replace('\n','').strip())
         
        
        #CTC and Date
        money=soup.find_all('div',class_='item_body')
        c=2
        for i in money:
            if c%3==0:
                ctc.append(i.text.replace('\n','').strip())
            elif c%3==1:
                date.append(i.text)
            c+=1
            
    #creating a dataframe
    df=pd.DataFrame()
    df['Job Title']=title
    df['Company']=company
    df['CTC']=ctc 
    df['Apply By']=date
    
    return df
        
```

```python
df=fresherjobs('https://internshala.com/fresher-jobs{}')
df
```
